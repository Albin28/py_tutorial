```python
income = int(input("Enter your income: "))
tax_rate = int(input("Enter tax rate (%): "))
tax = income * (tax_rate / 100)
print(f"Tax to be paid: {tax:.2f}")

```

```python
amount = float(input("Enter initial investment: "))
rate = float(input("Enter annual interest rate (%): "))
years = int(input("Enter number of years: "))

print("Year\tAmount")
for year in range(1, years+1):
    amount = amount + (amount * rate / 100)
    print(f"{year}\t{amount:.2f}")
```

```python
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num+1):
    fact *= i
print(f"Factorial of {num} is {fact}")
```

```python
n = int(input("Enter number of terms: "))
a, b = 0, 1
print("Fibonacci sequence:")
for _ in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

```python
num = int(input("Enter a number: "))
sum_val = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum_val += digit ** 3
    temp //= 10

if num == sum_val:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")
```

```python
num = float(input("Enter a number: "))
guess = num / 2
for _ in range(20):  
    guess = (guess + num / guess) / 2
print(f"Square root of {num} ≈ {guess:.5f}")
```

```python
m = int(input("Enter start of range: "))
n = int(input("Enter end of range: "))

print(f"Prime numbers between {m} and {n}:")
for num in range(m, n+1):
    if num > 1:
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

```python
def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

message = input("Enter message: ")
shift = int(input("Enter shift value: "))

encrypted = caesar_cipher(message, shift)
decrypted = caesar_cipher(encrypted, -shift)

print("Encrypted:", encrypted)
print("Decrypted:", decrypted)
```

```python
binary = input("Enter a binary number: ")
decimal = 0
for i, digit in enumerate(binary[::-1]):
    decimal += int(digit) * (2 ** i)
print(f"Decimal equivalent: {decimal}")
```

```python
string = input("Enter a string: ")


mid = len(string) // 2
if string[:mid] == string[mid:]:
    print("The string is symmetrical")
else:
    print("The string is not symmetrical")


if string == string[::-1]:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```
